```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

Notebook utilzzato per realizzare i grafici dell'addestramento dei modelli

# Modello 1

```python
with open('/content/drive/MyDrive/finetuned_model_logs.json', 'r') as f:
  logs_dict = json.load(f)
```

```python
train_loss = []
val_loss = []
val_acc = []

for i in range(len(logs_dict)-1):
  if i % 2 == 0:
    train_loss.append(logs_dict[i]['loss'])
  else:
    val_loss.append(logs_dict[i]['eval_loss'])
    val_acc.append(logs_dict[i]['eval_accuracy'])
```

```python
epochs = range(1, len(val_acc)+1)

plt.figure(figsize=(8,6))
plt.plot(epochs, train_loss, label='Training Loss', c='red')
plt.plot(epochs, val_loss, label='Validation Loss', c='blue')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(fontsize='large')
plt.axvline(x=np.argmin(val_loss)+1, c='black', linestyle='--', linewidth=2)
plt.text(x=np.argmin(val_loss)+1.1, y=0.4, s=f'Stopped at Epoch {np.argmin(val_loss)+1}', fontsize='large')
plt.text(x=np.argmin(val_loss)+1.1, y=0.37, s=f'Validation Loss: {round(val_loss[np.argmin(val_loss)], 2)}', c='blue', fontsize='large')
plt.text(x=np.argmin(val_loss)+1.1, y=0.34, s=f'Training Loss: {round(train_loss[np.argmin(val_loss)], 2)}', c='red', fontsize='large')
plt.xticks(epochs)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(epochs, val_acc, label='Validation Accuracy', c='blue')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(fontsize='large')
plt.axvline(x=np.argmax(val_acc)+1, c='black', linestyle='--', linewidth=2)
plt.text(x=np.argmax(val_acc)+1.1, y=0.687, s=f'Stopped at Epoch {np.argmax(val_acc)+1}', fontsize='large')
plt.text(x=np.argmax(val_acc)+1.1, y=0.6855, s=f'Validation Accuracy: {round(val_acc[np.argmax(val_acc)], 2)}', c='blue', fontsize='large')
plt.xticks(epochs)
plt.show()
```
